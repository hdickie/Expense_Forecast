```r
library(ggplot2)
library(reshape2)
```

```r
data <- read.table('/home/hdickie/sandbox/results.csv',sep=',',stringsAsFactors=FALSE,header=TRUE)
melted.data <- melt(data)

title_text=paste("Expense Forecast:",min(data$Date),'->',max(data$Date),collapse=' ')
subtitle_text=paste("Generated",Sys.time(),collapse=' ')

png('/home/hdickie/sandbox/results.png',width=1800,height=1000)

ggplot(melted.data,aes(x=Date,y=value,group=variable,color=variable)) + geom_line(size=2) + theme(plot.subtitle = element_text(hjust = 0.5,size=20), legend.title=element_text(size=14,face="bold"),axis.title=element_text(size=18,face="bold"), plot.title = element_text(hjust = 0.5,size=30),axis.text.y = element_text(size=18),axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1,size=18),legend.text=element_text(size=18)) + labs(title = title_text, subtitle = subtitle_text) 

dev.off()
```
